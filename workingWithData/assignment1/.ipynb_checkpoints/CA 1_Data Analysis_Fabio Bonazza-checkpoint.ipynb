### Data Processing  <a name="Data_Processing"></a>

```python
# loading in the Data from csv

# file = 'C:/Users/bonazza_fa/OneDrive - AXA/TU/Data Analysis/CA1_09112022/exchange_rate.csv' # Fabio work pc
file = 'C:/Users/Fabio/Downloads/exchange_rate.csv' # Fabio personal pc

# Read csv file
df = pd.read_csv(file)

# Display top 5 rows
df.head()
```

```python
# Transpose the dataset
df_T = df.transpose()

# Descibe the columns type
df_T.info() 
```

```python
# Use the 1st row as column names
df_T = df_T.rename(columns=df_T.iloc[0]).iloc[1:]

# Displat the top 5 rows
df_T.head()
```

```python
# Change all the dataset type from str to float
df_T = df_T.astype(float)

# Display the database basic info
df_T.info()
```

```python
# Index as 1st column
df_T.reset_index(inplace = True)

# Rename index column to 'Date' 
df_T.rename(columns={'index':'Date'}, inplace=True)

# Displat the top 5 rows
df_T.head()
```

```python
# Check the column 'Date' type 
df_T[['Date']].info()
```

```python
# Change the date new type from sting to date
df_T['Date'] = pd.to_datetime(df_T['Date'], format='%Y-%m-%d')

# Check the column type
df_T[['Date']].info()
```

```python
# Investigate all the elements whithin each Feature 
for column in df_T:
    unique_vals = np.unique(df_T[column])
    nr_values = len(unique_vals)
    if nr_values < 10:
        print('The number of unique values for feature {} :{} -- {}'.format(column, nr_values,unique_vals))
    else:
        print('The number of unique values for feature {} :{}'.format(column, nr_values))
```

```python
# Investigate all the elements with less then 2 values
for column in df_T:
    unique_vals = np.unique(df_T[column])
    nr_values = len(unique_vals)
    if nr_values < 2:
        print('Thecurrency {} has a fixed correlation with USD with the value of: -- {}'.format(column,unique_vals))
```

```python
# Loading in the Bitcoin Data form yahoo finance (https://finance.yahoo.com/quote/BTC-USD/history/)
print('Loading last 2 year price of Bitcoin:')
bicoin_price = yf.download(tickers='BTC-USD', period='2y', interval='1d')

# Change the first column from index to column and rename it as date
bicoin_price = bicoin_price.reset_index()
bicoin_price = bicoin_price.rename(columns = {"index" : "Date"})
bicoin_price = bicoin_price.rename(columns = {"Datetime" : "Date"})
bicoin_price['Date'] = pd.to_datetime(bicoin_price['Date'], format='%Y-%m-%d')

# Print the dataframe shape
print('\nRows, Columns: ',bicoin_price.shape)

# Runs the last 5 rows
bicoin_price.tail()
```

```python
# Check the df data types
bicoin_price.info()
```

```python
# Create a new colum inverse BTC-USD, to get how many bitcoin you can get with a US dollar
bicoin_price['Close_inverse'] = 1/bicoin_price['Close'] 

# Display the 1st df row
bicoin_price.head(1)
```

```python
# Create a new dataset with two columns Date
bitcoin = bicoin_price[['Date','Close_inverse']]

# Runs the last 5 rows
bitcoin.tail()
```

```python
# Rename Close column to 'Date' 
bitcoin.rename(columns={'Close_inverse':'BTC'}, inplace=True)

# Displat the top 5 rows
bitcoin.head()
```

```python
# Join the two dataframes with left join
prices = df_T.merge(bitcoin, on='Date', how='left', indicator=False)

# Display the database basic info
prices.info()
```

```python
# Display all the column names
prices.columns
```

```python
# Check the price trend for each currency
for i in prices.columns[1:]:
    prices.plot(x='Date', y=i)
```

```python
# Statistical Info from numeric variables
prices.describe().T
```

```python
# Create a new dataframe with the 1st and last price for each currency and pivot 
price_diff = pd.concat([prices.tail(1), prices.head(1)],ignore_index=True).T

# Use the 1st row as column names
price_diff = price_diff.rename(columns=df_T.iloc[0]).iloc[1:]

# Rename columns 
price_diff.rename(columns={0:'Start',1:'End'}, inplace=True)

# Add a Trend column
price_diff.loc[price_diff['End'] > price_diff['Start'],'Trend'] = 'Positive'
price_diff.loc[price_diff['End'] < price_diff['Start'],'Trend'] = 'Negative'
price_diff.loc[price_diff['End'] == price_diff['Start'],'Trend'] = 'Costant'
price_diff

# Create the percentage difference column between the end value and the first
price_diff['%_Diff'] = (price_diff['End'] / price_diff['Start']-1)*100

# Sort values by the '%_Diff' column
price_diff.sort_values(['%_Diff'], inplace = True)

# Display dataset
price_diff
```

```python
# Statistical Info from numeric variables
price_diff.describe().T
```

```python
# List of currency where USD has a negative trend
price_diff[price_diff['Trend'] == 'Negative']
```

```python
# Create the average value for EUR
eur_mean = prices['EUR'].mean()

# Plot the EUR time series
prices.plot(x='Date', y='EUR', figsize=(12,8))

# Add the EUR average to the chart
plt.axhline(y=eur_mean, color='red', linestyle='--', linewidth=3, label='Avg')

# Display all legends
plt.legend()

plt.title('EUR and yealy Average')

# Display the chart with all added features
plt.show()
```

```python
# Moving Average 10 days and 50 Days

# Create a new dataset from prices
eur_ts = prices[['Date','EUR']]

# Add column 10 days MA
eur_ts['MA10'] = eur_ts['EUR'].rolling(10).mean()

# Add column 50 days MA
eur_ts['MA50'] = eur_ts['EUR'].rolling(50).mean()

# Change fig size
fig, ax = plt.subplots(figsize=(12,8))

# Create a subplot in seaborn
ax= sns.lineplot(x='Date', y='EUR', data=eur_ts)
ax1 = sns.lineplot(x='Date', y='MA10', data=eur_ts)
ax2 = sns.lineplot(x='Date', y='MA50', data=eur_ts)

# Add legend on subplot in seaborn
ax.legend(['EUR', 'MA10','MA50'])

# Add Title on subplot in seaborn
ax.title.set_text('EUR Vs. MA10 Vs. MA50')
```
